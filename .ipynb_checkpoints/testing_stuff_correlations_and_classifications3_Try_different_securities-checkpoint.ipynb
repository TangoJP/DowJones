```python
dow = ra.ReturnAnalysis('./stocks/^DJI.csv', windows=[1, 5, 10, 20, 50, 100, 200])
sp500 = ra.ReturnAnalysis('./stocks/^GSPC.csv', windows=[1, 5, 10, 20, 50, 100, 200])
amzn = ra.ReturnAnalysis('./stocks/AMZN.csv', windows=[1, 5, 10, 20, 50, 100, 200])
axp = ra.ReturnAnalysis('./stocks/AXP.csv', windows=[1, 5, 10, 20, 50, 100, 200])
fb = ra.ReturnAnalysis('./stocks/FB.csv', windows=[1, 5, 10, 20, 50, 100, 200])
ge = ra.ReturnAnalysis('./stocks/^DJI.csv', windows=[1, 5, 10, 20, 50, 100, 200])
intc = ra.ReturnAnalysis('./stocks/GE.csv', windows=[1, 5, 10, 20, 50, 100, 200])
jpm = ra.ReturnAnalysis('./stocks/INTC.csv', windows=[1, 5, 10, 20, 50, 100, 200])
mmm = ra.ReturnAnalysis('./stocks/JPM.csv', windows=[1, 5, 10, 20, 50, 100, 200])
utx = ra.ReturnAnalysis('./stocks/MMM.csv', windows=[1, 5, 10, 20, 50, 100, 200])
vz = ra.ReturnAnalysis('./stocks/VZ.csv', windows=[1, 5, 10, 20, 50, 100, 200])

data = [dow, sp500, amzn, axp, fb, ge, intc, jpm, mmm, utx, vz]
labels = ['dow', 'sp500', 'amzn', 'axp', 'fb', 'ge', 'intc', 'jpm', 'mmm', 'utx', 'vz']
```

```python
fig = plt.figure(figsize=(12, 32))
for i, company in enumerate(data):
    corr = company.current_future_correlations()
    ax = fig.add_subplot(6, 2, i+1)
    sns.heatmap(corr, vmin=-1, vmax=1, annot=True, ax=ax)
    ax.set_title(labels[i], fontsize=14)
plt.tight_layout()
```
