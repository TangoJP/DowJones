```python
import os
import pandas as pd
import numpy as np
from matplotlib import cm
import matplotlib.pyplot as plt
import seaborn as sns
import stock
import base_analysis
import return_analysis as ra
import itertools
import scipy.stats as scs
from importlib import reload
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
dows = {}
dir_path = './stocks/dow_components'
component_file_list = os.listdir(dir_path)
for filename in component_file_list:
    file_path = dir_path + '/' + filename
    name = filename.replace('.csv', '')
    dows[name] = ra.ReturnAnalysis(file_path, windows=[1, 5, 10, 20, 50, 100, 200])
```

```python
len(dows)
```

```python
fig = plt.figure(figsize=(16, 50))
i = 1
for k, v in dows.items():
    corr = v.current_future_correlations()
    ax = fig.add_subplot(10, 3, i)
    sns.heatmap(corr, vmin=-1, vmax=1, annot=True, ax=ax)
    ax.set_title(k, fontsize=14)
    i += 1
plt.tight_layout()
```

So it appears that the past long term returns (LTRs)(i.e. retunr up to the current date) correlate with the LTRs in the future more strongly for some stocks, but not others. However, such correlations are very weak at best, although the results indicate some degree of self-correlation for some stocks.
