```python
# SMs & EMs
window = 100
span = 100
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(12, 6))
intc.plotSMA(ax=ax1, window=window)
intc.plotSMVar(ax=ax2, window=window)
intc.plotSMStd(ax=ax3, window=window)
intc.plotEMA(ax=ax4, span=span)
intc.plotEMVar(ax=ax5, span=span)
intc.plotEMStd(ax=ax6, span=span)
plt.tight_layout()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 3))
intc.daily_return_hist(scale='log', ax=ax1)
intc.daily_return_hist(scale='linear', ax=ax2)
plt.tight_layout()
```

```python
vz = stock.Stock('VZ')
```

```python
# SMs & EMs
window = 100
span = 100
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(12, 6))
vz.plotSMA(ax=ax1, window=window)
vz.plotSMVar(ax=ax2, window=window)
vz.plotSMStd(ax=ax3, window=window)
vz.plotEMA(ax=ax4, span=span)
vz.plotEMVar(ax=ax5, span=span)
vz.plotEMStd(ax=ax6, span=span)
plt.tight_layout()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 3))
vz.daily_return_hist(scale='log', bins=60, ax=ax1)
vz.daily_return_hist(scale='linear', bins=60, ax=ax2)
plt.tight_layout()
```
